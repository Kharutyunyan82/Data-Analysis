```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
'''Use NumPy to generate random data'''
import numpy as np
x = np.linspace(0, 10, 25)
y = x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2))25
```

```python
'''It only takes 1 command to draw'''
pylab.plot(x,y, 'r')
```

```python
'''Drawing a subgraph
'''
pylab.subplot(1,2,1) # The contents of the request(rows, columns, indexes)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8], facecolor='y')
axis.plot(x,y, 'r')
```

```python
# again we can draw subgraphs
fig, axes = plt.subplots(nrows=1, ncols=3) # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# we can also draw a picture, or graph, inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3]) # Small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi=300) # new graphic object
fig.add_subplot(321)
plt.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots(dpi=300)
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('title of graph')
axes.plot(x,y, 'r')
axes.plot(x, x**3)
axes.legend(['y = x**2', 'y = x**3'], loc=2)
```

# Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])
fig, axes = plt.subplots(1, 4, figsize=(16,5))
axes[0].set_title('scatter')
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n, n**2, lw=2)

axes[2].set_title('bar')
axes[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)

axes[3].set_title('fill_between')
axes[3].fill_between(n, n**2, n**3, color='green', alpha=0.5)
```

```python
'''Draw a radar chart
'''
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t= np.linspace(0, 2*np.pi, 100)
ax.plot(t,t*0.5, color='blue', lw=3)
```

```python
'''Draw a histogram
'''
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title('Default histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram')
axes[1].hist(n, cumulative=True, bins=50)
```

```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm #colormap
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X,Y = np.meshgrid(x, y)
Z1 = np.exp((-X**2 - Y**2))
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

print(X)
print('\n')
print(Z)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline = 1, fontsize=10)
ax.set_title('contour map')
```

```python
# Draw 3D surface image
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14, 6), dpi=120)

# Specify the 3D graphics to draw, with projection = '3d'
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
```

## Write a Python programming to create a pie chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
#Data to plot
languages = 'Python', 'Java', 'PHP', 'Javascript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
```

```python
# explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, explode=explode, labels=languages, colors=colors, 
        autopct = '%1.1f%%', shadow=True, startangle=140, radius=2)
```
