```python
# import the necessary Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Load the car.csv dataset
car = pd.read_csv(r"C:\Users\karen\Desktop\Data_Analysis_Projects\Data_Analysis\Data Analysis using Python\car.csv")
```

```python
# Relationship between categorical and continious variable
sns.boxplot(x='Vehicle Size', y='Engine HP', data=car)
```

```python
sns.pairplot(car)
```

```python
car.columns
```

```python
# Drop irrelevant columns
car = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
car.head(6)
```

```python
# Renaming the column names.
car = car.rename(columns={'Engine HP': 'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission', 'Driven_Wheels': 'Drive Mode', 'highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price'})
car.head(6)
```

```python
# Total number of rows and columns
car.shape
```

```python
# Rows containing duplicate data
duplicate_rows_car = car[car.duplicated()]
print('Number of duplicate rows: ', duplicate_rows_car.shape)
```

```python
# Count the rows before removing the data
car.count()
```

```python
# Drop the duplicates
car = car.drop_duplicates()
car.head()
```

```python
# Find the null values
print(car.isnull().sum())
```

```python
# Drop the missing values
car = car.dropna()
car.count()
```

```python
# Finding the outliers
sns.boxplot(x=car['Price'])
```

```python
sns.boxplot(x=car['HP'])
```

```python
# Plotting a histogram for number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(16, 8))
```

```python
# Find relation between the variables
plt.figure(figsize=(20, 10))
c= car.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```
